```python
# Plotting Functions
def plot_line_chart(dataset_name, data):
    if 'Close' in data.columns:
        plt.figure(figsize=(12, 6))
        plt.plot(data.index, data['Close'], label='Close', color='blue', linewidth=1.5)
        plt.plot(data.index, data['10-Day MA'], label='10-Day MA', color='green', linestyle='--')
        plt.plot(data.index, data['50-Day MA'], label='50-Day MA', color='orange', linestyle='--')
        plt.title(f"{dataset_name} - Line Chart", fontsize=16)
        plt.xlabel('Date', fontsize=12)
        plt.ylabel('Price', fontsize=12)
        plt.legend()
        plt.tight_layout()
        plt.show()
    else:
        print(f"Dataset {dataset_name} is missing required columns.")

def plot_area_chart(dataset_name, data):
    if all(col in data.columns for col in ['Close', '10-Day MA', '50-Day MA']):
        plt.figure(figsize=(12, 6))
        plt.fill_between(data.index, data['Close'], label='Close', color='blue', alpha=0.3)
        plt.fill_between(data.index, data['10-Day MA'], label='10-Day MA', color='green', alpha=0.3)
        plt.fill_between(data.index, data['50-Day MA'], label='50-Day MA', color='orange', alpha=0.3)
        plt.title(f"{dataset_name} - Area Chart", fontsize=16)
        plt.xlabel('Date', fontsize=12)
        plt.ylabel('Price', fontsize=12)
        plt.legend()
        plt.tight_layout()
        plt.show()
    else:
        print(f"Dataset {dataset_name} is missing required columns.")

def plot_bar_chart(dataset_name, data):
    if 'Weekly Returns' in data.columns:
        plt.figure(figsize=(12, 6))
        plt.bar(data.index, data['Weekly Returns'], color='purple', alpha=0.7, label='Weekly Returns')
        plt.title(f"{dataset_name} - Weekly Returns", fontsize=16)
        plt.xlabel('Date', fontsize=12)
        plt.ylabel('Returns', fontsize=12)
        plt.axhline(y=0, color='black', linewidth=0.8, linestyle='--')
        plt.legend()
        plt.tight_layout()
        plt.show()
    else:
        print(f"Dataset {dataset_name} does not have 'Weekly Returns'.")

def plot_heatmap(datasets):
    combined_data = pd.DataFrame()
    for dataset_name, dataset in datasets.items():
        if 'Close' in dataset.columns:
            combined_data[dataset_name] = dataset['Close']
    correlation_matrix = combined_data.corr()
    plt.figure(figsize=(10, 8))
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
    plt.title('Correlation Heatmap Between Indices')
    plt.tight_layout()
    plt.show()

# Central Function
def central_plotting_function(plot_type, datasets):
    for dataset_name, data in datasets.items():
        print(f"Generating {plot_type} for {dataset_name}...")
        if plot_type == 'line_chart':
            plot_line_chart(dataset_name, data)
        elif plot_type == 'area_chart':
            plot_area_chart(dataset_name, data)
        elif plot_type == 'bar_chart':
            plot_bar_chart(dataset_name, data)
        elif plot_type == 'heatmap':
            plot_heatmap(datasets)
            break  # Heatmap is a single chart for all datasets combined
        else:
            print(f"Invalid plot type: {plot_type}")

# Example Usage
datasets = {
    'NIFTY 50': nifty50_data,
    'NIFTY REALTY': REALTY_data,
    'NIFTY MEDIA': MEDIA_data,
    'NIFTY COMMODITIES': COMMODITIES_data,
    'NIFTY ENERGY': ENERGY_data,
    'NIFTY METAL': METAL_data,
    'NIFTY INFRA': INFRA_data,
    'NIFTY AUTO': autonifty_data,
    'NIFTY IT': itnifty_data,
    'NIFTY FMCG': fmcgnifty_data,
    'NIFTY BANK': banknifty_data
}


# Call the central function with the desired plot type
central_plotting_function('line_chart', datasets)
```

```python
import matplotlib.pyplot as plt
from sklearn.linear_model import BayesianRidge
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split

def predict_multiple_indices(datasets, target_column='Close', test_size=0.2, random_state=42):
    results = {}

    for name, data in datasets.items():
        # Split the data into features (X) and target (y)
        X = data.drop(columns=[target_column])  # Features, excluding the target column
        y = data[target_column]  # Target variable, in this case 'Close'

        # Split the dataset into training and testing sets
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=random_state)

        # Train the model
        model = BayesianRidge()
        model.fit(X_train, y_train)

        # Predict the values
        y_pred = model.predict(X_test)

        # Calculate the Mean Squared Error (MSE)
        mse = mean_squared_error(y_test, y_pred)

        # Store the result in the dictionary
        results[name] = mse

        # Residual Plot (Prediction Error Plot)
        plt.figure(figsize=(10, 6))
        residuals = y_test - y_pred
        plt.scatter(y_test.index, residuals, color='green', label='Residuals', alpha=0.6)
        plt.axhline(y=0, color='black', linestyle='--', linewidth=2)
        plt.title(f'{name} - Residual Plot')
        plt.xlabel('Index')
        plt.ylabel('Residuals (Actual - Predicted)')
        plt.legend(loc='upper left')
        plt.grid(True)
        plt.show()



    return results

# Example usage with the provided datasets
datasets = {
    'NIFTY 50': nifty50_data,
    'NIFTY REALTY': REALTY_data,
    'NIFTY MEDIA': MEDIA_data,
    'NIFTY COMMODITIES': COMMODITIES_data,
    'NIFTY ENERGY': ENERGY_data,
    'NIFTY METAL': METAL_data,
    'NIFTY INFRA': INFRA_data,
    'NIFTY AUTO': autonifty_data,
    'NIFTY IT': itnifty_data,
    'NIFTY FMCG': fmcgnifty_data,
    'NIFTY BANK': banknifty_data
}


# Call the function with the datasets
results = predict_multiple_indices(datasets)
```
